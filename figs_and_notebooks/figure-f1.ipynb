```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
model_performances = pd.read_excel(r'./model-key-f1.xlsx')
```

```python
model_performances
```

```python
fig, axes = plt.subplots(figsize=(10, 5))
sns.set_theme(style="ticks")
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xticks(size = 15)
plt.yticks(size = 18)
plt.xlabel("Test F1 Score", weight = 'bold', size = 18)
plt.subplots_adjust(left = 0.18)
# plt.yticks(rotation=30)
sns.violinplot(data=model_performances, ax = axes, orient ='h', scale = 'width', inner = None)
plt.scatter(x = model_performances['299-AN'], y = ['299-AN'] * 10, color = 'white', alpha = 0.3, s = 80)
plt.scatter(x = model_performances['Raw-EfficientNet-B3'], y = ['Raw-EfficientNet-B3'] * 10, color = 'white', alpha = 0.3, s = 80)
plt.scatter(x = model_performances['Pre-Trained-EfficientNet-B3'], y = ['Pre-Trained-EfficientNet-B3'] * 10, color = 'white', alpha = 0.3, s = 80)
plt.scatter(x = model_performances['Raw-VGG-16'], y = ['Raw-VGG-16'] * 10, color = 'white', alpha = 0.3, s = 80)
plt.scatter(x = model_performances['Pre-Trained-VGG-16'], y = ['Pre-Trained-VGG-16'] * 10, color = 'white', alpha = 0.3, s = 80)
plt.scatter(x = model_performances['Raw-ResNet-50'], y = ['Raw-ResNet-50'] * 10, color = 'white', alpha = 0.3, s = 80)
plt.scatter(x = model_performances['Pre-Trained-ResNet-50'], y = ['Pre-Trained-ResNet-50'] * 10, color = 'white', alpha = 0.3, s = 80)
plt.scatter(x = model_performances['Raw-ViT-L-16'], y = ['Raw-ViT-L-16'] * 10, color = 'white', alpha = 0.3, s = 80)
plt.scatter(x = model_performances['Pre-Trained-ViT-L-16'], y = ['Pre-Trained-ViT-L-16'] * 10, color = 'white', alpha = 0.3, s = 80)
# plt.violinplot(model_performances, vert = False)
plt.grid()
plt.savefig('relative-model-performances-f1.png', dpi = 1200, bbox_inches = 'tight')
```
